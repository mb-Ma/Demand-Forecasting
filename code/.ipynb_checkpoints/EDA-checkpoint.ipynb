```python
import pandas as pd
import numpy as np
```

```python
using_hist = pd.read_csv('../data/Dataset/demand_train_A.csv')
using_future = pd.read_csv('../data/Dataset/demand_test_A.csv')
inventory = pd.read_csv('../data/Dataset/inventory_info_A.csv')
last_dt = pd.to_datetime("20210301")
start_dt = pd.to_datetime("20210302")
end_dt = pd.to_datetime("20210607")
lead_time = 14
```

```python
using_hist["ts"] = pd.to_datetime(using_hist["ts"])
using_future["ts"] = pd.to_datetime(using_future["ts"])

qty_using = pd.concat([using_hist, using_future])
date_list = pd.date_range(start=start_dt, end=end_dt)
unit_list = using_future["unit"].unique()
res = pd.DataFrame(columns=["unit", "ts", "qty"])
```

```python
replenishUnit_dict = {}  # 补充量
demand_dict = {} # 需求量
```

```python
%matplotlib inline
for idx, chunk in enumerate(qty_using.groupby("unit")):
    unit = chunk[0]
    demand = chunk[1]
    demand.sort_values("ts", inplace=True, ascending=True)
    demand["diff"] = demand["qty"].diff().values
    demand["qty"] = demand["diff"]
    del demand["diff"]
    demand = demand[1:]
    
    if idx == 1:
        demand.sort_values("ts", inplace=True, ascending=True)
        demand.set_index("ts")["qty"].plot(title=unit)
        break
    
```
